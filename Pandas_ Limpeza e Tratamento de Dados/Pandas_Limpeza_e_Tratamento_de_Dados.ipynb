##Identificando outliers

Valor atípico, ou ponto fora da curva, em estatística o termo é usado para se referenciar a um dado que se diferencia muito da distribuição padrão

Boxplot é uma caixa onde no meio dela há a mediana, que divide os dados no meio
<link>https://www.youtube.com/watch?v=qU2lANG4hYQ<link>

Screenshot 2024-09-03 at 16-53-33 Pandas limpeza e tratamento de dados Aula 4 - Atividade 2 Identificando os outliers Alura - Cursos online de tecnologia.png


```python
df_sem_nulos.describe()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=df_sem_nulos['cliente.tempo_servico']);
```

```python
Q1 = df_sem_nulos['cliente.tempo_servico'].quantile(0.25) #selecionando quartil
Q3 = df_sem_nulos['cliente.tempo_servico'].quantile(0.75)
IQR = Q3 - Q1 #intervalo interquartil
# (r = range)
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
outliers_index = (df_sem_nulos['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulos['cliente.tempo_servico'] > limite_superior)
```

```python
df_sem_nulos[outliers_index]['cliente.tempo_servico']
```

##Substituindo valores para os outliers

Tratamos outliers para evitar que o modelo machine learning aprenda com base em padrões incorretos

```python
df_sem_outliers = df_sem_nulos.copy()
```

```python
df_sem_outliers[outliers_index]['cliente.tempo_servico']
```

```python
df_sem_outliers.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    df_sem_outliers.loc[outliers_index,'conta.cobranca.Total'] /
    df_sem_outliers.loc[outliers_index,'conta.cobranca.mensal']
    )
```

```python
sns.boxplot(x=df_sem_outliers['cliente.tempo_servico'])
```

```python
df_sem_outliers[outliers_index][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

###Removendo outliers



```python
df_sem_outliers[outliers_index]['cliente.tempo_servico']
```

```python
#refazendo o filtro

Q1 = df_sem_outliers['cliente.tempo_servico'].quantile(0.25) #selecionando quartil
Q3 = df_sem_outliers['cliente.tempo_servico'].quantile(0.75)
IQR = Q3 - Q1 #intervalo interquartil
# (r = range)
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers_index = (df_sem_outliers['cliente.tempo_servico'] < limite_inferior) | (df_sem_outliers['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
df_sem_outliers[outliers_index]
```

```python
df_sem_outliers = df_sem_outliers[~outliers_index]
                                  #todos as amostras menos os outliers
df_sem_outliers
```

```python
sns.boxplot(x=df_sem_outliers['cliente.tempo_servico']);
```

```python
df_sem_outliers.reset_index(drop=True, inplace=True)
#resetando os index
```

```python
df_sem_outliers
```

#Trabalhando com variáveis categóricas

###Substituindo valores
variável categórica, pois o valor que é atribuído a ela não é numérico.

Elas são forma de agrupar informações em categorias distintas

Variáveis categóricas binárias possuem apenas duas categorias distintas

```python
df_sem_id = df_sem_outliers.drop('id_cliente', axis=1).copy()
```

```python
#variáveis categóricas binárias
mapeamento = {
    'sim': 1,
    'nao': 0,
    'masculino': 0,
    'feminino' : 1
}

for col in df_sem_id.columns:
  print(f'Coluna: {col}')
  print(df_sem_id[col].unique())
  print('-'*30)
```

```python
colunas = ['telefone.servico_telefone', 'Churn', 'cliente.parceiro', 'cliente.dependentes', 'conta.faturamente_eletronico', 'cliente.genero']
```

```python
df_sem_id[colunas] = df_sem_id[colunas].replace(mapeamento)
df_sem_id
```

```python
for col in df_sem_id.columns:
  print(f'Coluna: {col}')
  print(df_sem_id[col].unique())
  print('-'*30)
```
