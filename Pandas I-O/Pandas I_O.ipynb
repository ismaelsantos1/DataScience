##Criando um banco de dados local

### SQL (Structed Query Language)

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, text ,MetaData, Table, inspect #permite inspecionar elementos do bd
                      #criar a conexão com o banco de dados
```

```python
engine = create_engine('sqlite:///:memory:') #cria um banco de dados em memória (local)
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False) #salva no formato para o banco de dados
```

```python
inspect(engine).get_table_names() #mostra as tabelas do banco de dados
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
#vai selecionar todas as colunas da tabela clientes onde todas
                                          # as pessoas que tem categoria de renda == empregado
        #cláusulas são usadas em caixa alta
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados.head()
```

```python
empregados.to_sql('empregados', con= engine, index=False)
                                #usado para especificar o objeto do sql alchemy
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

####Atualizando banco de dados

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
#with serve para garantir que a conexão seja fechada corretamente após a consulta
    #connnect se liga ao banco de dados e execute executa a instrução
with engine.connect() as conn:
    conn.execute(text(query))
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = 'UPDATE clientes SET Grau_escolaridade = "Ensino superior" WHERE ID_Cliente= 5008808'
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit() #para atualizar os dados
```

```python
pd.read_sql_table('clientes', engine)
```
