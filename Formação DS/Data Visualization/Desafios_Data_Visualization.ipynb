#Aula 1

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Alura/DataScience/Formação DS/Data Visualization/imigrantes_canada.csv')
df.set_index('País', inplace=True)
anos = list(map(str, range(1980, 2014)))
```

```python
import matplotlib.pyplot as plt
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.plot(df_comparacao['Brasil'], label= 'Brasil')
plt.plot(df_comparacao['Argentina'], label='Argentina')
plt.title("Imigração do Brasil e Argentina para o Canadá\n1980 a 2013")
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

#Aula 2

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.title('Vendas do ano de 2022 nas lojas A, B, C e D')

# Adicionar os gráficos em cada um dos subplots
axs[0, 0].plot(df.loc['A'])
axs[0, 0].set_title('Vendas na loja A')
axs[0, 1].plot(df.loc['B'])
axs[0, 1].set_title('Vendas na loja B')
axs[1, 0].plot(df.loc['C'])
axs[1, 0].set_title('Vendas na loja C')
axs[1, 1].plot(df.loc['D'])
axs[1, 1].set_title('Vendas na loja D')

for ax in axs.flat:
  ax.set(xlabel='Mês', ylabel='Número de vendas')

plt.show()

```
