```python
import pandas as pd

# Conjunto de dados TechTaste
df_techtaste = pd.DataFrame({'avaliacoes': [38, 44, 33, 42, 47, 33, 36, 39, 42, 36, 39, 34, 42, 42, 36, 43, 31, 35, 36, 41, 42, 30, 25, 38, 47, 36, 32, 45, 44, 45, 37, 48, 37, 36, 44, 49, 31, 45, 45, 40, 36, 50, 38, 34, 36, 42, 46, 49, 36, 34, 38, 31, 53, 40, 57, 40, 36, 42, 26, 50, 32, 43, 35, 37, 42, 30, 36, 43, 40, 43, 44, 52, 37, 51, 35, 47, 40, 50, 37, 49]})
```

```python
#values previamente especificados
desvio_padrao_pop = 2.65
confianca = 0.95

#outros valores para calcular o IC
tamanho_amostra = len(df_techtaste)
media_nula = 30

#media amostral
media_amostral = df_techtaste['avaliacoes'].mean()
print(f'Média Amostral: {media_amostral}')

'''
Hipóteses:

H0: μ = 30
H1: μ ≠ 30
'''

#calculando intervalo de confianca
ic = stats.norm.interval(confianca,
                       loc=media_nula,
                       scale=(desvio_padrao_pop)/np.sqrt(tamanho_amostra))
print(f'Intervalo de Confiança ({confianca*100}%): {ic}')

'''
Média amostral fora do intervalo de confiança, assim a H0 é rejeitada
'''
```

```python
#Teste Z para calcular o valor da Z-Score e o p-valor para o problema da TechTaste
from statsmodels.stats.weightstats import ztest

stats, p_valor = ztest(x1= df_techtaste['avaliacoes'], value=30,
                       alternative='two-sided')

print(f'Z-Score: {stats}')
print(f'p-valor: {p_valor}')
```

```python
nivel_significancia = 0.05

if p_valor < nivel_significancia:
  print(f'Hipótese nula rejeitada')
else:
  print(f'Hipótese nula não rejeitada')
```

```python
import pandas as pd

df_equipe_vendas = pd.DataFrame({'Vendedor': [ 'Luíza', 'Bia', 'Rodrigo', 'Allan', 'Evaldo'],
                                 'Vendas Antes (R$)': [252.72, 203.91, 307.32, 185.78, 220.5],
                                 'Vendas Depois (R$)': [285.1, 223.15, 324.41, 202.23, 240.63]})
```

```python
'''
1. Amostra  dependente
2.
  H0: μ das diferenças das observações é igual a 0
  H1: μ das diferenças das observações é diferente de 0
'''

resultado_subtracao = [b - a for a, b in zip(df_equipe_vendas['Vendas Antes (R$)'],
                       df_equipe_vendas['Vendas Depois (R$)'])]
resultado_subtracao
```

```python
from scipy.stats import ttest_rel
stats, p_valor = ttest_rel(df_equipe_vendas['Vendas Depois (R$)'], df_equipe_vendas['Vendas Antes (R$)'])

print('P_Valor: ', p_valor*100,'%')
print('Stats: ', stats)
```

```python
nivel_significancia = 0.05

if p_valor < nivel_significancia:
  print(f'Hipótese nula rejeitada:  O resultado do p-valor foi menor que o nível de significância,\n por isso rejeitamos a hipótese nula.\n O novo treinamento gerou uma diferença significativa nas vendas.')
else:
  print(f'Hipótese nula não rejeitada')
```
