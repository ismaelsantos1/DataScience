```python
 dados.info()
```

dados tipo object pode ser dado geralmente por ter vários tipos diferentes numa coluna.
Outra forma de aparecer é quando os dados tem um tipo específico, mas o tipo não foi especificado e aí o próprio pandas atribui esse dtype. Além disso, também poderia mostrar object em caso de todos os tipos serem strings



```python
import numpy as np
```

```python
#tratando dados numéricos - convertendo
#astype converte coluna para um tipo específico de dado
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
#convertendo para tipo float
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info() #conferindo a conversão
```

```python
#tratando a coluna preco
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())#strip remove qqlr caractere antes ou depois
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']]= dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

```python
#tokenização é a divisão de um texto em partes menores
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
#regex é um expressão regular que se caracteriza pela pesquisa no texto
#busca | validacao | manipulacao
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
                                                                #selecione todos exceto esses
dados['descricao_local']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
                                                            #verifica se tem algum caractere antes ou depois
                                                            #se não tiver, remove o hífen
dados['descricao_local']
```

```python
#tokenização de strings
dados['descricao_local'] = dados['descricao_local'].str.split()
dados['descricao_local']
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].replace('(?<!\w)-(?!\w)',' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados['descricao_vizinhanca']
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
                                                  #remove chaves e aspas
dados['comodidades']
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
